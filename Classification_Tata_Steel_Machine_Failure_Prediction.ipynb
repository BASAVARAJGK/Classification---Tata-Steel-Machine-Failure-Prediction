## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant difference in torque between failed and non-failed machines.

Alternative Hypothesis (H1): There is a significant difference in torque between failed and non-failed machines.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

fail = df[df['Machine_failure']==1]['Torque_Nm']
not_fail = df[df['Machine_failure']==0]['Torque_Nm']

t_stat, p_value = ttest_ind(fail, not_fail)

print("p-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Independent T-Test

Used for:
Torque vs Machine Failure

##### Why did you choose the specific statistical test?

Independent T-Test
The Independent T-Test is used to compare the means of a numerical variable between two independent groups.
In this project:
Numerical features (e.g., Torque, Tool Wear)
Two groups:
Machine Failure = 0
Machine Failure = 1

👉 Therefore, T-Test is appropriate to check whether the difference in means is statistically significant.

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

Technique Used: Class Weighting

The dataset is highly imbalanced, with machine failure cases representing only a small percentage (~1.6%) of the total data. To address this issue, class weighting was used in the XGBoost model.

The parameter scale_pos_weight was set based on the ratio of majority to minority class, allowing the model to give more importance to the minority class (machine failure).

🔹 Why this technique was chosen?
It preserves the original data distribution without removing or duplicating samples.
It reduces the risk of overfitting compared to oversampling techniques.
It avoids information loss associated with undersampling.
It is highly effective for tree-based models like XGBoost, which support class weighting directly.

Other techniques considered
Undersampling was not used as it would discard valuable data from the majority class.
